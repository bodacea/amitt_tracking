```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd

gsheeturl = "https://docs.google.com/spreadsheets/d/1pmEqn1rgGjzSIynpVupPySdnBOdqUbOaWJpH6yKFBrI/edit#gid=335661034"

# use the Google Drive API to grab the googlesheet
scope = ['https://spreadsheets.google.com/feeds']
creds = ServiceAccountCredentials.from_json_keyfile_name('/Users/sara/gdrive_key_mispload-e665c0d21327.json', scope)
client = gspread.authorize(creds)
doc = client.open_by_url(gsheeturl)
sheetnames = [x.title for x in doc.worksheets()]
print('Spreadsheet: {}\nWorksheets: {}\n'.format(doc.title, ', '.join(sheetnames)))
```

```python
# Extract values from objects worksheet
sheet = doc.worksheet('objects')
list_of_hashes = sheet.get_all_records()
df = pd.DataFrame(list_of_hashes)
df = df[df['to_object'].str.len() > 0]
df
```
