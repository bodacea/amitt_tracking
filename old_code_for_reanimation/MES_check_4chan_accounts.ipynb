# Check 4chan accounts on youtube

```python
import pandas as pd

df = pd.read_csv('../data/4chan_coalpatrol_youtubeaccounts.csv', 
                 header=None, names=['URL'])
df = df.drop_duplicates()

dfx = df['URL'].str.split('/', expand=True)
dfx.drop(axis=0, columns=[0,1,2], inplace=True)
dfx.rename(columns={3:'type', 4:'name', 5:'featured'}, inplace=True)
dfx
```

```python
dfx[dfx['type'] == 'user'].sort_values('name')
```

```python
dfx[dfx['type'] == 'channel'].sort_values('name')
```
