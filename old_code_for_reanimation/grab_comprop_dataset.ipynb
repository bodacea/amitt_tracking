# Grab comprop dataset

```python
import pandas as pd

df = pd.read_excel('2018_08_comprop/Seed_List_18.xlsx')
df
```

```python
import requests

for nr, url in zip(df['nr'].tolist(), df['decoded'].tolist()):

    try:
        r = requests.get(url, allow_redirects=True)
        open('2018_08_comprop/seed_list_18/{}.html'.format(nr), 'wb').write(r.content)
        print('downloaded {}'.format(url))
    except:
        print('Couldn\'t download {}'.format(url))
```
