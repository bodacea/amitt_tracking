```python
from twarc import Twarc
import pandas as pd

# Set up link to Twitter
secretfile='/Users/sara/twittersecrets.txt'
with open(secretfile, 'r') as fsecret: 
  secrets = fsecret.readline()
access_token, access_token_secret, consumer_key, consumer_secret = \
    [x.strip() for x in secrets.split(',')]
twarc = Twarc(consumer_key, consumer_secret, access_token, access_token_secret)
```

```python
# Read data from your CSV file - put your own details in here
# usercsv = pd.read_csv('userdeets.csv')
# usernames = usercsv['usernames'].tolist()
usernames = ['bodaceacat', 'xxevgnjb3342', 'rungrage'] #comment this out

allusers = []
for user in twarc.user_lookup(usernames, id_type='screen_name'):
    allusers += [user]
df = pd.DataFrame(allusers)
df
```

```python
df[['screen_name', 'created_at']].to_csv('names_with_creation_dates.csv', index=False)
df[['screen_name', 'created_at']].sort_values('created_at', ascending=False)
```
