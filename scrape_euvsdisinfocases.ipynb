# scrape EUvsDisinfo dataset

Looking for these: div class='disinfo-db-post'

```python
#import scrapy #thinking about running a scrapy spider https://scrapy.org/
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
import os


if not os.path.exists('data/scrapes'):
    os.makedirs('data/scrapes')

if not os.path.exists('data/scrapes/euvsdisinfopages'):
    os.makedirs('data/scrapes/euvsdisinfopages')

baseurl = 'https://euvsdisinfo.eu/disinformation-cases/?offset={}'

html = requests.get(baseurl.format(0)).text
soup = bs(html, 'lxml')
rescount = int(soup.find('div', 'disinfo-db-results').find('span').text)
npages = int(np.ceil(rescount/10.0))
print('{} entries in database'.format(rescount))
```

```python
allposts = []
for page in range(npages):
    print(page)
    html = requests.get(baseurl.format(10*page)).text
    soup = bs(html, 'lxml')
    posts = soup.find_all('div', 'disinfo-db-post')
    # date, title, outlets, country, url
    for i, post in enumerate(posts): 
        pdata = [i,
                 post.find(attrs={'data-column':'Date'}).text.strip(),
                 post.find(attrs={'data-column':'Title'}).text.strip(),
                 post.find(attrs={'data-column':'Outlets'}).text.strip(),
                 post.find(attrs={'data-column':'Country'}).text.strip(),
                 post.find(attrs={'data-column':'Title'}).find('a').get('href')]
        allposts += [pdata]
df = pd.DataFrame(allposts, columns=['Index', 'Date', 'Title', 'Outlets', 'Country', 'NATO_url'])
df.to_csv('data/scrapes/euvsdisinfoheadings.csv', index=False)
df
```

## Now get into the individual pages

```python
allpages = []
for i, url in enumerate(df['NATO_url'].to_list()):
    if int(i/10) == i/10:
        print('{} {}'.format(i, url))

    try:
        html = requests.get(url).text
        with open('data/scrapes/euvsdisinfopages/{}.txt'.format(url[url.strip('/').rfind('/'):].strip('/')), 'w') as page_file:
            page_file.write(html)
    except:
        print('failed writing {}'.format(url))

    try:
        soup = bs(html, 'lxml')
        bodyclasses = soup.find('body').get('class')
        pdata = [url,
                 bodyclasses[[i for i, s in enumerate(bodyclasses) if s.startswith('postid-')][0]],
                 soup.find('h1', 'b-catalog__report-title').text,
                 soup.find('div', 'b-report__summary-text').text.strip(),
                 soup.find('div', 'b-report__disproof-text').text.strip()]
        if soup.find('div', 'b-catalog__repwidget').find('a') != None:
            pdata += [[(x.text, x.get('href')) for x in soup.find('div', 'b-catalog__repwidget').find_all('a')]]
        else:
            pdata += [[]]
        pdata += [[bb.text.strip() for bb in soup.find('ul', 'b-catalog__repwidget-list').find_all('li')]]
    except:
        print('Scrape failed for {}'.format(url))
        pdata = [url]
    allpages += [pdata]

dfpages = pd.DataFrame(allpages, columns=['NATO_url', 'Id', 'PageTitle', 'Summary', 'Disproof', 'Media', 'Details'])
dfpages.to_csv('data/scrapes/dfpages.csv', index=False)
dfpages
```
