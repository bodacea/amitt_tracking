```python
import pandas as pd

scrapes = [
    '20201012112116_blexit',
    '20201012115030_walkaway',
    '20201012115258_fba',
    '20201012115352_blackvoicesfortrump',
    '20201012115545_ados',
    '20201012115545_american_descendents_of_slaves',
    '20201012154951_black_voices_for_trump',
    '20201012161739_foundational_black_americans'    
]

df = pd.DataFrame([])
template = '../data/twitter/{0}/{0}_tweets.csv'

for i, name in enumerate(scrapes):
    dfin = pd.read_csv(template.format(name))
    dfin['source'] = name
    df = df.append(dfin)

df.to_csv('../data/all_blackgroupdata.csv', index=False)
df
```
