```python
# How many tweets have we double-counted? Turns out not that many.  Only 11 are in 3 places; 1199 are in 2. 
dfrepeats = df['tweet_url'].value_counts().reset_index()
dfrepeats['tweet_url'].value_counts()
```

```python
dfrepeated = df[df['tweet_url'].isin(dfrepeats[dfrepeats['tweet_url']>2]['index'])]
dfrepeated
```
